# 치사율 데이터 가져오기

#### 데이터 소스: https://github.com/owid/covid-19-data/tree/master/public/data

```python
data = pd.read_csv("./data/owid-covid-data.csv")
data
```

```python
deathData = data.total_deaths
caseData = data.total_cases
```

```python
countrySeries = data.iso_code
countryList = list(set(countrySeries)) # set으로 만들어서 중복 다 없앰.

len(countryList)
```

### 국가 filtering하기

```python
countryRowList = []
for country in countryList:
    check = False
    for i in range(len(countrySeries)):
        if(country == countrySeries[i]):
            check = True
        if(check and country != countrySeries[i]): # 맨 마지막 total값을 가져오기 위해서 country[i]이름이 달라지면 i-1을 리스트에 넣는다.
            if(caseData[i-1] > 3000 and country != 'OWID_WRL'): # case가 3000명이 넘는 것만 들고 옴, OWID_WRL은 전세계 데이터라 뺌
                countryRowList.append(i-1)
            break
            
len(countryRowList) 
```

### {country: fatality rate} 형태로 dictionary만들기

```python
fatalityRateDict = {}
for index in countryRowList:
    if(not math.isnan(deathData[index]) and not math.isnan(caseData[index])):
        fatalityRateDict[countrySeries[index]] = 100*float(deathData[index]/caseData[index])
```

```python
print(len(fatalityRateDict))
fatalityRateDict # 치사율 데이터
```

# Factor 데이터 dictionary로 바꾸기

#### 데이터 소스: https://github.com/owid/covid-19-data/tree/master/public/data

```python
data = pd.read_csv("./data/owid-covid-data.csv")
data
```

### Factor Data 고르기

```python
ageOver65 = data.aged_65_older
ageOver70 = data.aged_70_older
```

```python
ageOver65Count = 0
ageOver70Count = 0
for index in countryRowList:
    if(not math.isnan(ageOver65[index])):
        ageOver65Count +=1
    if(not math.isnan(ageOver70[index])):
        ageOver70Count +=1
        
print("The number of female smokers data: {}".format(ageOver65Count))
print("The number of male smokers data: {}".format(ageOver70Count))
```

### {Country: Data}인 dictionary 만들기

```python
ageOver70Dict = {}
ageOver65Dict = {}
for index in countryRowList:
    if(not math.isnan(ageOver65[index])): # nan 아닌 데이터들만 사용
        ageOver65Dict[countrySeries[index]] = ageOver65[index]
    if(not math.isnan(ageOver70[index])):
        ageOver70Dict[countrySeries[index]] = ageOver70[index]
```

```python
ageOver65Dict 
```

# Correlation 구하기

### 비교할 리스트 만들어 주기

```python
fatalityListforageOver65 = []
fatalityListforageOver70 = []
ageOver65List = []
ageOver70List = []

for isoCode in fatalityRateDict:
    if isoCode in ageOver65Dict:
        fatalityListforageOver65.append(fatalityRateDict[isoCode])
        ageOver65List.append(ageOver65Dict[isoCode])
    if isoCode in ageOver70Dict:
        fatalityListforageOver70.append(fatalityRateDict[isoCode])
        ageOver70List.append(ageOver70Dict[isoCode])
```

```python
print("correlation list for ageOver65: {} : {}".format(len(fatalityListforageOver65), len(ageOver65List)))
print("correlation list for ageOver70: {} : {}".format(len(fatalityListforageOver70), len(ageOver70List)))
```

```python
correlationAgeOver65 = [fatalityListforageOver65, ageOver65List]
correlationAgeOver70 = [fatalityListforageOver70, ageOver70List]
```

### correlation 구하기

```python
df_ageOver65 = pd.DataFrame(correlationAgeOver65).T
corr_ageOver65 = df_ageOver65.corr(method= 'pearson')
print("Correlation between age Over 65 and fatality: {}".format(corr_ageOver65[1][0]))

df_ageOver70 = pd.DataFrame(correlationAgeOver70).T
corr_ageOver70 = df_ageOver70.corr(method= 'pearson')
corr_ageOver70[1][0]
print("Correlation between age Over 70 and fatality: {}".format(corr_ageOver70[1][0]))
```
