```python
import pandas as pd
import math
import collections
```

# 치사율 데이터 가져오기

#### 데이터 소스: https://github.com/owid/covid-19-data/tree/master/public/data

```python
data = pd.read_csv("./data/owid-covid-data.csv")
data
```

```python
deathData = data.total_deaths
caseData = data.total_cases
```

```python
countrySeries = data.iso_code
countryList = list(set(countrySeries)) # set으로 만들어서 중복 다 없앰.

len(countryList)
```

### 국가 filtering하기

```python
countryRowList = []
for country in countryList:
    check = False
    for i in range(len(countrySeries)):
        if(country == countrySeries[i]):
            check = True
        if(check and country != countrySeries[i]): # 맨 마지막 total값을 가져오기 위해서 country[i]이름이 달라지면 i-1을 리스트에 넣는다.
            if(caseData[i-1] > 3000 and country != 'OWID_WRL'): # case가 3000명이 넘는 것만 들고 옴, OWID_WRL은 전세계 데이터라 뺌
                countryRowList.append(i-1)
            break
            
len(countryRowList) 
```

### {country: fatality rate} 형태로 dictionary만들기

```python
fatalityRateDict = {}
for index in countryRowList:
    if(not math.isnan(deathData[index]) and not math.isnan(caseData[index])):
        fatalityRateDict[countrySeries[index]] = 100*float(deathData[index]/caseData[index])
```

```python
print(len(fatalityRateDict))
fatalityRateDict # 치사율 데이터
```

# Factor 데이터 dictionary로 바꾸기

#### 데이터 소스: https://data.oecd.org/air/air-pollution-exposure.htm

```python
data = pd.read_csv("./data/공기오염/Fine-particulate-matter(PM2.5).csv")
data
```

```python
data_countrySeries = data.LOCATION
data_countryList = list(set(data.LOCATION))
```

### Factor Data 고르기

```python
airPollution = data.Value
len(airPollution)
```

```python
data_countryRowList = []
for country in data_countryList:
    check = False
    for i in range(len(data_countrySeries)):
        if(country == data_countrySeries[i]):
            check = True
        if(check and country != data_countrySeries[i]): # 맨 마지막 total값을 가져오기 위해서 country[i]이름이 달라지면 i-1을 리스트에 넣는다.
            if(country != 'OECDE'): # case가 3000명이 넘는 것만 들고 옴, OWID_WRL은 전세계 데이터라 뺌
                data_countryRowList.append(i-1)
            break
            
len(data_countryRowList) 
```

### {Country: Data}인 dictionary 만들기

```python
airPollutionDict = {}
for index in data_countryRowList:
    if(not math.isnan(airPollution[index])): # nan 아닌 데이터들만 사용
        airPollutionDict[data_countrySeries[index]] = airPollution[index]
```

```python
airPollutionDict
```

# Correlation 구하기

### 비교할 리스트 만들어 주기

```python
fatalityList = []
airPollutionList = []

for isoCode in fatalityRateDict:
    if isoCode in airPollutionDict:
        fatalityList.append(fatalityRateDict[isoCode])
        airPollutionList.append(airPollutionDict[isoCode])
```

```python
print("correlation list for air pollution {} : {}".format(len(fatalityList), len(airPollutionList)))
```

```python
correlationAirPollution = [fatalityList, airPollutionList]
```

### correlation 구하기

```python
df = pd.DataFrame(correlationAirPollution).T
corr = df.corr(method= 'pearson')
print("Correlation between air pollution and fatality: {}".format(corr[1][0]))
```
