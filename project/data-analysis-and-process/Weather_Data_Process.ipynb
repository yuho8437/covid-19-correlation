```python
import pandas as pd

data = pd.read_csv("./data/날씨/daily_weather_2020.csv", index_col=None)
data = data[["Country/Region", "time", "sunriseTime","sunsetTime", "precipIntensity", "pressure", "windSpeed", "ozone","temperatureHigh", "temperatureLow", 'humidity']]
```

```python
UIDtoCountry = pd.read_csv("./data/UIDtoISO.csv")

UIDtoCountry
```

```python
UIDtoCountryList = UIDtoCountry[["Country_Region", "iso3"]].values.tolist()

UIDtoCountryList
```

```python
dataList = data.values.tolist()
dataList
```

```python
for row in dataList:
    for convert in UIDtoCountryList:
        if(row[0] == convert[0]):
            row[0] = convert[1]
            break
```

```python
newDataList = []
doneList = []
for row in dataList:
    if([row[0],row[1]] in doneList):
        continue
    count = 1
    daytime = float(row[3]) - float(row[2])
    precip = float(row[4])
    pressure =float(row[5])
    windspeed = float(row[6])
    ozone = float(row[7])
    tempHighTotal = float(row[8])
    tempLowTotal = float(row[9])
    humidityTotal = float(row[10])
    for row2 in dataList:
        if(row[0] == row2[0] and row[1] == row2[1]):
            count+=1
            daytime += float(row[3]) - float(row[2])
            precip += float(row[4])
            pressure +=float(row[5])
            windspeed += float(row[6])
            ozone += float(row[7])
            tempHighTotal+= float(row2[8])
            tempLowTotal += float(row2[9])
            humidityTotal += float(row2[10])
    doneList.append([row[0],row[1]])
    newDataList.append([row[0], row[1], daytime/count, precip/count, pressure/count, windspeed/count, ozone/count, (tempHighTotal+tempLowTotal)/(2*count), humidityTotal/count])
```

```python
newDataList
```

```python
df = pd.DataFrame(newDataList, columns = ["Country", "Date","Daytime", "Precip", "Pressure", "Windspeed", "Ozone", "Temperature", "Humidity"])
```

```python
df.to_csv('new_weather_Processed_Data.csv', index=False)
```
