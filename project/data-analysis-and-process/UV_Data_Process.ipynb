```python
import pandas as pd

data = pd.read_csv("./data/uvbed.csv/uvbed.csv")
data
```

```python
UIDtoISO = pd.read_csv("./data/UIDtoISO.csv")

UIDtoISOList = UIDtoISO[["UID", "iso3"]].values.tolist()

UIDtoISOList
```

```python
dataList = data.values.tolist()
```

```python
for row in dataList:
    for convert in UIDtoISOList:
        if(row[0] == convert[0]):
            row[0] = convert[1]
            break
```

```python
dataList
```

```python
newDataList = []
doneList = []
for row in dataList:
    if([row[0],row[1]] in doneList):
        continue
    count = 1
    total = float(row[2])
    for row2 in dataList:
        if(row[0] == row2[0] and row[1] == row2[1]):
            count+=1
            total+= float(row2[2])
    doneList.append([row[0],row[1]])
    newDataList.append([row[0], row[1], total/count])
```

```python
df = pd.DataFrame(newDataList, columns = ["Country", "Date", "UV"])
```

```python
df.to_csv('list.csv', index=False)
```
