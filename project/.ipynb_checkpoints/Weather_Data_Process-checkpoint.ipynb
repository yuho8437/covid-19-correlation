```python
import pandas as pd

data = pd.read_csv("./data/날씨/daily_weather_2020.csv", index_col=None)
data = data[["Country/Region", "time", "temperatureHigh", "temperatureLow", 'humidity']]
```

```python
UIDtoCountry = pd.read_csv("./data/UIDtoISO.csv")

UIDtoCountry
```

```python
UIDtoCountryList = UIDtoCountry[["Country_Region", "iso3"]].values.tolist()

UIDtoCountryList
```

```python
dataList = data.values.tolist()
dataList
```

```python
for row in dataList:
    for convert in UIDtoCountryList:
        if(row[0] == convert[0]):
            row[0] = convert[1]
            break
```

```python
newDataList = []
doneList = []
for row in dataList:
    if([row[0],row[1]] in doneList):
        continue
    count = 1
    tempHighTotal = float(row[2])
    tempLowTotal = float(row[3])
    humidityTotal = float(row[4])
    for row2 in dataList:
        if(row[0] == row2[0] and row[1] == row2[1]):
            count+=1
            tempHighTotal+= float(row2[2])
            tempLowTotal += float(row2[3])
            humidityTotal += float(row2[4])
    doneList.append([row[0],row[1]])
    newDataList.append([row[0], row[1], (tempHighTotal+tempLowTotal)/(2*count), humidityTotal/count])
```

```python
newDataList
```

```python
df = pd.DataFrame(newDataList, columns = ["Country", "Date", "Temperature", "Humidity"])
```

```python
df.to_csv('weather_Processed_Data.csv', index=False)
```
