```python
import pandas as pd
import math
```

# 치사율 데이터 가져오기

#### 데이터 소스: https://github.com/owid/covid-19-data/tree/master/public/data

```python
data = pd.read_csv("./data/owid-covid-data.csv")
data
```

```python
deathData = data.total_deaths
caseData = data.total_cases
```

```python
countrySeries = data.iso_code
countryList = list(set(countrySeries)) # set으로 만들어서 중복 다 없앰.

len(countryList)
```

```python
countryRowList = []
for country in countryList:
    check = False
    for i in range(len(countrySeries)):
        if(country == countrySeries[i]):
            check = True
        if(check and country != countrySeries[i]):
            countryRowList.append(i)
            break
            
len(countryRowList) # 한 개가 모자라는데, 그 한개는 nan임.
```

```python
fatalityRateDict = {}
for index in countryRowList:
    if(not math.isnan(deathData[index]) and not math.isnan(caseData[index])):
        if(caseData[index] == 0):
            fatalityRateDict[countrySeries[index]] = 0
        else:
            fatalityRateDict[countrySeries[index]] = deathData[index]/caseData[index]
```

# Factor 데이터 dictionary로 바꾸기

#### 데이터 소스: https://github.com/owid/covid-19-data/tree/master/public/data

```python
data = pd.read_csv("./data/owid-covid-data.csv")
data
```

### Factor Data 고르기

```python
femaleSmokers = data.female_smokers
maleSmokers = data.male_smokers
len(countrySeries)
```

```python
femaleCount = 0
maleCount = 0
for index in countryRowList:
    if(not math.isnan(femaleSmokers[index])):
        femaleCount +=1
    if(not math.isnan(maleSmokers[index])):
        maleCount +=1
        
print("The number of female smokers data: {}".format(femaleCount))
print("The number of male smokers data: {}".format(maleCount))
```

### {Country: Data}인 dictionary 만들기

```python
countryMaleSmokerDict = {}
countryFemaleSmokerDict = {}
for index in countryRowList:
    if(not math.isnan(femaleSmokers[index])): # nan 아닌 데이터들만 사용
        countryFemaleSmokerDict[countrySeries[index]] = femaleSmokers[index]
    if(not math.isnan(maleSmokers[index])):
        countryMaleSmokerDict[countrySeries[index]] = maleSmokers[index]
```
