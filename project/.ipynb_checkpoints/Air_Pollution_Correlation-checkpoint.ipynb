```python
import pandas as pd
import math
import collections
```

# 치사율 데이터 가져오기

#### 데이터 소스: https://github.com/owid/covid-19-data/tree/master/public/data

```python
data = pd.read_csv("./data/owid-covid-data.csv")
data
```

```python
deathData = data.total_deaths
caseData = data.total_cases
```

```python
countrySeries = data.iso_code
countryList = list(set(countrySeries)) # set으로 만들어서 중복 다 없앰.

len(countryList)
```

### 국가 filtering하기

```python
countryRowList = []
for country in countryList:
    check = False
    for i in range(len(countrySeries)):
        if(country == countrySeries[i]):
            check = True
        if(check and country != countrySeries[i]): # 맨 마지막 total값을 가져오기 위해서 country[i]이름이 달라지면 i-1을 리스트에 넣는다.
            if(caseData[i-1] > 3000 and country != 'OWID_WRL'): # case가 3000명이 넘는 것만 들고 옴, OWID_WRL은 전세계 데이터라 뺌
                countryRowList.append(i-1)
            break
            
len(countryRowList) 
```

### {country: fatality rate} 형태로 dictionary만들기

```python
fatalityRateDict = {}
for index in countryRowList:
    if(not math.isnan(deathData[index]) and not math.isnan(caseData[index])):
        fatalityRateDict[countrySeries[index]] = 100*float(deathData[index]/caseData[index])
```

```python
print(len(fatalityRateDict))
fatalityRateDict # 치사율 데이터
```

# Factor 데이터 dictionary로 바꾸기

#### 데이터 소스: https://github.com/owid/covid-19-data/tree/master/public/data

```python
data = pd.read_csv("./data/owid-covid-data.csv")
data
```

### Factor Data 고르기

```python
femaleSmokers = data.female_smokers
maleSmokers = data.male_smokers
len(countrySeries)
```

```python
femaleCount = 0
maleCount = 0
for index in countryRowList:
    if(not math.isnan(femaleSmokers[index])):
        femaleCount +=1
    if(not math.isnan(maleSmokers[index])):
        maleCount +=1
        
print("The number of female smokers data: {}".format(femaleCount))
print("The number of male smokers data: {}".format(maleCount))
```

### {Country: Data}인 dictionary 만들기

```python
maleSmokerDict = {}
femaleSmokerDict = {}
for index in countryRowList:
    if(not math.isnan(femaleSmokers[index])): # nan 아닌 데이터들만 사용
        femaleSmokerDict[countrySeries[index]] = femaleSmokers[index]
    if(not math.isnan(maleSmokers[index])):
        maleSmokerDict[countrySeries[index]] = maleSmokers[index]
```

```python
maleSmokerDict 
```

# Correlation 구하기

### 비교할 리스트 만들어 주기

```python
fatalityListforFemale = []
fatalityListforMale = []
femaleSmokerList = []
maleSmokerList = []

for isoCode in fatalityRateDict:
    if isoCode in femaleSmokerDict:
        fatalityListforFemale.append(fatalityRateDict[isoCode])
        femaleSmokerList.append(femaleSmokerDict[isoCode])
    if isoCode in maleSmokerDict:
        fatalityListforMale.append(fatalityRateDict[isoCode])
        maleSmokerList.append(maleSmokerDict[isoCode])
```

```python
print("correlation list for female {} : {}".format(len(fatalityListforFemale), len(femaleSmokerList)))
print("correlation list for male {} : {}".format(len(fatalityListforMale), len(maleSmokerList)))
```

```python
correlationSmokingFemale = [fatalityListforFemale, femaleSmokerList]
correlationSmokingMale = [fatalityListforMale, maleSmokerList]
```

### correlation 구하기

```python
df_female = pd.DataFrame(correlationSmokingFemale).T
corr_female = df_female.corr(method= 'pearson')
print("Correlation between female smokers and fatality: {}".format(corr_female[1][0]))

df_male = pd.DataFrame(correlationSmokingMale).T
corr_male = df_male.corr(method= 'pearson')
corr_male[1][0]
print("Correlation between male smokers and fatality: {}".format(corr_male[1][0]))
```
