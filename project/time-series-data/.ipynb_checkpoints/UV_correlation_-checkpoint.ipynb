# 국가, 날짜별 치사율 데이터 가져오기

#### 데이터 소스: https://github.com/owid/covid-19-data/tree/master/public/data

```python
data = pd.read_csv("../data/owid-covid-data.csv")
data
```

### 국가 filtering하기

```python
countrySeries = data.iso_code
countryList = list(set(countrySeries)) # set으로 만들어서 중복 다 없앰.

len(countryList)
```

```python

caseData = data.total_cases
```

```python
selectedCountries = []
countryRowList = []
for country in countryList:
    check = False
    for i in range(len(countrySeries)):
        if(country == countrySeries[i]):
            check = True
        if(check and country != countrySeries[i]): # 맨 마지막 total값을 가져오기 위해서 country[i]이름이 달라지면 i-1을 리스트에 넣는다.
            if(caseData[i-1] > 3000 and country != 'OWID_WRL'): # case가 3000명이 넘는 것만 들고 옴, OWID_WRL은 전세계 데이터라 뺌
                countryRowList.append(i-1)
                selectedCountries.append(country)
            break
            
len(countryRowList) 
```

```python
selectedCountries
```

### 일별 치사율 구하기

```python

data = data[["iso_code",'date', 'total_cases', 'total_deaths']] 
fatalityDataList = data.values.tolist()
```

```python
fatalityData = []
for row in fatalityDataList:
    if(row[0] in selectedCountries):
        if(row[2] != 0):
            row.append(100* (row[3]/row[2]))
        else:
            row.append(0)

        del row[3]
        del row[2]
        fatalityData.append(row)
```

```python
fatalityData # 맨 마지막 column이 치사율임
```

# Factor 데이터 time-series로 바꾸기

#### 데이터 소스: https://www.kaggle.com/vishalvjoseph/weather-dataset-for-covid19-predictions

```python
data = pd.read_csv("../data/날씨/weather_Processed_Data.csv")
data
```

### 누적데이터 일별 데이터로 바꾸기

```python
temp = data[['Country', 'Date','Temperature']]
humidity = data[['Country', 'Date','Humidity']]
tempList = temp.values.tolist()
humidityList = humidity.values.tolist()
```

```python
tempList
```

### 나라별로 데이터 묶어서 Correlation구하기

#### 온도 Correlation 구하기

```python
factorNumList = tempList

corrDict = {}
factorSeries = []
fatalitySeries = []
curCountry = ''

for rowF in fatalityData:
    prevCountry = curCountry
    curCountry = rowF[0]
    if(prevCountry != curCountry and prevCountry != ''):
        corrList = [factorSeries, fatalitySeries]
        df = pd.DataFrame(corrList).T
        corr = df.corr(method= 'pearson')

        corrDict[prevCountry] = corr[1][0]
        factorSeries = []
        fatalitySeries = []
    
    for rowT in factorNumList:
        if(rowT[0] == rowF[0] and rowT[1] == rowF[1]):
            factorSeries.append(rowT[2])
            fatalitySeries.append(rowF[2])
```

```python
print("온도 Correlation: ")
corrDict
```

### 평균 Correlation구하기

```python
count = 0
total = 0
for elem in list(corrDict.values()):
    if not math.isnan(elem):
        count +=1
        total += elem
total/count
```

#### 습도 Correlation 구하기

```python
factorNumList = humidityList

corrDict = {}
factorSeries = []
fatalitySeries = []
curCountry = ''

for rowF in fatalityData:
    prevCountry = curCountry
    curCountry = rowF[0]
    if(prevCountry != curCountry and prevCountry != ''):
        corrList = [factorSeries, fatalitySeries]
        df = pd.DataFrame(corrList).T
        corr = df.corr(method= 'pearson')

        corrDict[prevCountry] = corr[1][0]
        factorSeries = []
        fatalitySeries = []
    
    for rowT in factorNumList:
        if(rowT[0] == rowF[0] and rowT[1] == rowF[1]):
            factorSeries.append(rowT[2])
            fatalitySeries.append(rowF[2])
```

```python
print("습도 Correlation: ")
corrDict
```

```python
count = 0
total = 0
for elem in list(corrDict.values()):
    if not math.isnan(elem):
        count +=1
        total += elem
total/count
```
